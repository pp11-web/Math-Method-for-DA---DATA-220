```python
# Load the breast cancer dataset
cancer = load_breast_cancer()
X, y = cancer.data, cancer.target
```

```python
# Split the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=2023)
```

```python
# Train a linear regression model
linreg = LinearRegression()
linreg.fit(X_train, y_train)

# Predict on the test set
y_pred = linreg.predict(X_test)

# Convert predictions to binary (0 or 1)
y_pred_binary = [1 if pred >= 0.5 else 0 for pred in y_pred]
```

```python
# Compute mean squared error (MSE)
mse = mean_squared_error(y_test, y_pred_binary)

print(f'Mean Squared Error: {mse}')
```

```python
from sklearn.preprocessing import StandardScaler

# Standardizing the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

n_iterations = 100
learning_rate = 0.001
loss_history = []
# Linear Regression using Gradient Descent with feature scaling
def linear_regression(X, y, learning_rate, n_iterations):
    intercept = np.random.normal()
    coefficients = np.random.normal(size=X.shape[1])
    m = len(y)

    for _ in range(n_iterations):
        y_pred = intercept + np.dot(X, coefficients)
        loss = mean_squared_error(y, y_pred)

        # Check for numerical overflow
        if np.isinf(loss) or np.isnan(loss):
            return intercept, coefficients, float('inf')

        # Gradient calculation
        intercept_gradient = -2 * np.sum(y - y_pred) / m
        coefficients_gradient = -2 * np.dot(X.T, y - y_pred) / m

        # Update parameters
        intercept -= learning_rate * intercept_gradient
        coefficients -= learning_rate * coefficients_gradient
        
        loss_history.append(loss)
    return intercept, coefficients, loss

# Training the model with scaled features
optimal_intercept, optimal_coefficients, final_loss = linear_regression(X_train_scaled, y_train, learning_rate, n_iterations)

# Print the optimal parameters and the final loss
print("Optimal Intercept:", optimal_intercept)
print("Optimal Coefficients:", optimal_coefficients)
print("Final Loss:", final_loss)
```
